```python
saved_rm = open_play_shots_rm[open_play_shots_rm["shot_outcome"] == "Saved"]
saved_rm
```

### Shot Map Open Play Real Madrid

```python
from mplsoccer import VerticalPitch
pitch = VerticalPitch(pitch_type='statsbomb', half=True, goal_type='box', goal_alpha=0.8, pitch_color='grass', line_color='#c7d5cc')
fig, axs = pitch.grid(figheight=10, title_height=0.08, endnote_space=0, axis=False,title_space=0, grid_height=0.82, endnote_height=0.05)
fig.set_facecolor('#ffff')

# map shot
scatter_shots = pitch.scatter(open_play_shots_rm.x_location, open_play_shots_rm.y_location, 
                              s=(open_play_shots_rm.shot_statsbomb_xg * 900) + 100, c='red', edgecolors='black', marker='o', ax=axs['pitch'])
scatter_shots_end = pitch.scatter(open_play_shots_rm.x_end_location, open_play_shots_rm.y_end_location, 
                                  s=(open_play_shots_rm.shot_statsbomb_xg * 900) + 100, c='blue', edgecolors='black', marker='o', ax=axs['pitch'])

# map shot goal
scatter_location_goals = pitch.scatter(goals_rm.x_location, goals_rm.y_location, 
                                       s=(goals_rm.shot_statsbomb_xg * 900) + 100, c='red', edgecolors='black', marker='*', ax=axs['pitch'])
scatter_location_end_goals = pitch.scatter(goals_rm.x_end_location, goals_rm.y_end_location, 
                                           s=(goals_rm.shot_statsbomb_xg * 900) + 100, c='blue', edgecolors='black', marker='*', ax=axs['pitch'])

# map shots block
scatter_shots_blocked = pitch.scatter(blocked_rm.x_location, blocked_rm.y_location, 
                                      s=(blocked_rm.shot_statsbomb_xg * 900) + 100, c='yellow', edgecolors='black', marker='+', ax=axs['pitch'])
scatter_shots_blocked_end_location = pitch.scatter(blocked_rm.x_end_location, blocked_rm.y_end_location, 
                                                   s=(blocked_rm.shot_statsbomb_xg * 900) + 100, c='yellow', edgecolors='black', marker='+', ax=axs['pitch'])

# map shots saved
scatter_shots_saved = pitch.scatter(saved_rm.x_location, saved_rm.y_location, 
                                      s=(saved_rm.shot_statsbomb_xg * 900) + 100, c='yellow', edgecolors='black', marker='^', ax=axs['pitch'])
scatter_shots_saved_end_location = pitch.scatter(saved_rm.x_end_location, saved_rm.y_end_location, 
                                                   s=(saved_rm.shot_statsbomb_xg * 900) + 100, c='yellow', edgecolors='black', marker='^', ax=axs['pitch'])

axs['endnote'].text(0.85, 0.5, '[AWF01]', color='#c7d5cc', va='center', ha='center', fontsize=15)
axs['title'].text(0.5, 0.7, 'Shot Map Real Madrid Final Champions League 2018', color='#c7d5cc', va='center', ha='center', fontsize=30)
axs['title'].text(0.5, 0.25, 'Open Play | Full Time', color='#c7d5cc', va='center', ha='center', fontsize=18)
```

```python
fk_rm = real_madrid_1[real_madrid_1['pass_type'] == 'Free Kick']
fk_rm
```

```python
fk_rm['x_location'] = fk_rm.location.apply(lambda x: x[0])
fk_rm['y_location'] = fk_rm.location.apply(lambda x: x[1])
fk_rm['x_end_location'] = fk_rm.pass_end_location.apply(lambda x: x[0])
fk_rm['y_end_location'] = fk_rm.pass_end_location.apply(lambda x: x[1])
fk_rm = fk_rm.drop('location', axis=1)
fk_rm = fk_rm.drop('pass_end_location', axis=1)

fk_rm.head()
```

### Free Kick Maps Real Madrid 

```python
from mplsoccer import Pitch
pitch = Pitch(pitch_type='statsbomb', goal_type='box', goal_alpha=0.5, pitch_color='grass', line_color='#c7d5cc')
fig, axs = pitch.grid(figheight=10, title_height=0.08, endnote_space=0, axis=False,title_space=0, grid_height=0.82, endnote_height=0.05)
fig.set_facecolor('#ffff')

scatter_fk_pass_location = pitch.scatter(fk_rm.x_location, fk_rm.y_location, c='red', edgecolors='black', marker='o', ax=axs['pitch'])
scatter_fk_pass_location = pitch.scatter(fk_rm.x_end_location, fk_rm.y_end_location, c='blue', edgecolors='black', marker='o', ax=axs['pitch'])

axs['endnote'].text(0.85, 0.5, '[AWF01]', color='#c7d5cc', va='center', ha='center', fontsize=15)
axs['title'].text(0.5, 0.7, 'Free Kick Maps Real Madrid Final UCL 2018', color='#c7d5cc', va='center', ha='center', fontsize=30)
axs['title'].text(0.5, 0.25, 'Full Time', color='#c7d5cc', va='center', ha='center', fontsize=18)
```

### Free Kick Maps Real Madrid With Lines

```python
pitch = Pitch(pitch_type='statsbomb', pitch_color='grass', line_color='#c7d5cc')
fig, ax = pitch.draw(figsize=(10, 7), constrained_layout=True, tight_layout=False)
fig.set_facecolor('#ffff')

pitch.lines(fk_rm.x_location, fk_rm.y_location,
             fk_rm.x_end_location, fk_rm.y_end_location, lw=5, transparent=True, comet=True, color='#ba4f45', ax=ax)

axs['endnote'].text(0.85, 0.5, '[AWF01]', color='#c7d5cc', va='center', ha='center', fontsize=15)
axs['title'].text(0.5, 0.7, 'Free Kick Maps Real Madrid Final UCL 2018', color='#c7d5cc', va='center', ha='center', fontsize=30)
axs['title'].text(0.5, 0.25, 'Full Time', color='#c7d5cc', va='center', ha='center', fontsize=18)
```
